```python
import pandas as pd
```

```python
data_df = pd.read_csv('data/final-data.csv')
data_df.head()
```

```python
unique_uni_name_by_uni_code = data_df[['university_code', 'university_name']].groupby('university_code').nunique()
# Convert to Series
unique_uni_name_by_uni_code = unique_uni_name_by_uni_code['university_name']
unique_uni_name_by_uni_code
```

```python
unique_uni_name_by_uni_code[unique_uni_name_by_uni_code != 1]
```

```python
list_uni_name_by_code = data_df[['university_code', 'university_name']].groupby('university_code').agg(lambda x: list(set(x.tolist())))
list_uni_name_by_code = list_uni_name_by_code['university_name']
list_uni_name_by_code
```

```python
uni_code_to_name = list_uni_name_by_code.apply(lambda x: x[0]).to_dict()
uni_code_to_name
```

```python
new_uni_name = data_df[['university_code', 'university_name']]\
                .apply(lambda row: uni_code_to_name[row['university_code']], axis=1)
new_uni_name
```

```python
data_df['university_name'] = new_uni_name
```

```python
data_df = data_df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'Unnamed: 0.2'])
data_df
```

```python
data_df.to_csv('data/final-data.csv')
```
